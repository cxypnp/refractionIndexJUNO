Python Symbolic Calculation

```python
from sympy import *
from sympy.simplify.fu import *
init_printing()

A, P, r= symbols('A P r',real=true)
p, s = symbols('R_p R_s',real=1)

initVector=Matrix([1,r])
rotationP=Matrix([[cos(P),-sin(P)],[sin(P),cos(P)]])
reflection=Matrix([[p,0],[0,s]])
rotationA=Matrix([[cos(A),sin(A)],[-sin(A),cos(A)]])
polarizor=Matrix([[1,0],[0,r]])

finalVector=polarizor*rotationA*reflection*rotationP*initVector

finalVectorConjugate=(polarizor*rotationA*reflection.conjugate()*rotationP*initVector).T

intensity=(finalVectorConjugate*finalVector)[0,0].expand()
```

```python
result=collect(collect(TR7(TR5(TR7(trigsimp(intensity,method="combined")))).expand(),[cos(2*A),sin(2*A),cos(2*P),sin(2*P),r]).subs(sin(P)*cos(P),Rational(1, 2)*sin(2*P)),[cos(2*A),sin(2*A),r],evaluate=True)
result
```

```python
result=collect(collect(TR7(TR5(TR7(trigsimp(intensity,method="combined")))).expand(),[cos(2*A),sin(2*A),cos(2*P),sin(2*P),r]).subs(sin(P)*cos(P),Rational(1, 2)*sin(2*P)),[cos(2*A),sin(2*A),r],evaluate=False)
result
```

```python
constTerm=collect(result[1]+result[r]*r+result[r**2]*r**2+result[r**3]*r**3+result[r**4]*r**4,[cos(2*P),sin(2*P)])
constTerm
```

```python
sinTerm=collect(result[sin(2*A)],[cos(2*P),sin(2*P)])
sinTerm
```

```python
cosTerm=collect(result[cos(2*A)],[cos(2*P),sin(2*P)])
cosTerm
```

```python
finalModel=constTerm+sinTerm*sin(2*A)+cosTerm*cos(2*A)
finalModel
```

```python
result=collect(collect(TR7(TR5(TR7(trigsimp(intensity,method="combined")))).expand(),[cos(2*A),sin(2*A),cos(2*P),sin(2*P),r]).subs(sin(P)*cos(P),Rational(1, 2)*sin(2*P)),[cos(2*A),sin(2*A),r],evaluate=True)
result
```

```python
finalModel.compare(result)
```
